```python
import cv2

def convert_video_to_grayscale(input_video_path, output_video_path, resize_factor=None):
    # Open the input video
    cap = cv2.VideoCapture("/content/Bunny.mp4")

    # Check if the video opened successfully
    if not cap.isOpened():
        print("Error: Could not open the video.")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if resize_factor:
        frame_width = int(frame_width * resize_factor)
        frame_height = int(frame_height * resize_factor)

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height), isColor=False)

    frame_index = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Convert frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        if resize_factor:
            gray_frame = cv2.resize(gray_frame, (frame_width, frame_height))

        # Write the grayscale frame to the output video
        out.write(gray_frame)

        # Print processing status
        frame_index += 1
        print(f"Processed frame {frame_index}/{frame_count}")

    # Release resources
    cap.release()
    out.release()
    print("Conversion completed. Grayscale video saved as", output_video_path)

# Example usage
input_video_path = '/content/Bunny.mp4'
output_video_path = 'output_video_grayscale.mp4'
convert_video_to_grayscale(input_video_path, output_video_path, resize_factor=0.5)
```
